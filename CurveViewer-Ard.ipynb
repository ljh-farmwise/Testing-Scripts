```python
%matplotlib notebook

# import required modules
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import glob

# Select all CSV files on Desktop
files = glob.glob(r"\Users\Lawrence Hsiung\NEWtests1024\rr1\*.csv")
xaxis = np.linspace(0,1000, 10000)

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

plt.title('NEW Characterizations ')
i = 0
startpos = []
endpos = []
for file in files:
    if 'open-' in file:
        s_time = time.time()
        df = pd.read_csv(file, header = None)
        e_time = time.time()
        #print("Read time: ", (e_time-s_time), "seconds")
        df= df.T
        #print(df)
        df.columns =['Ch1', 'Ch2', 'Ch3', 'Ch4']
        zpt = sum(df['Ch3'][200:400])/len(df['Ch3'][200:400])
        df['Ch3'] = (df['Ch3']-zpt)/0.066666 #calibration factor
        df['Ch4'] = df['Ch4'] - 0.13
        
        if len(df['Ch3']) < 10005:
            ax1.plot(xaxis[4500:8500], df['Ch1'][4500:8500], label = 'INPUT %s' %i)
            ax2.plot(xaxis[4500:8500], df['Ch3'][4500:8500], label = '%s' %file[-22:])
            startpos.append(sum(df['Ch3'][2500:3500])/len(df['Ch3'][2500:3500]))
            endpos.append(sum(df['Ch3'][7500:8500])/len(df['Ch3'][7500:8500]))
            i+=1

plt.axvline(x=505, ls = ':', c = 'g')
plt.axvline(x=555, ls = ':', c = 'g')
plt.axvline(x=528, ls = ':', c = 'r')
# plt.axvline(x=57, ls = ':', c = 'r')
# plt.axvline(x=61, ls = ':', c = 'g')
ax1.set_xlabel('10*ms')
ax1.set_ylabel('Vsense', color='g')
ax2.set_ylabel('angle delta', color='b')
ax1.set_ylim([-0.5, 5])

#plt.legend()
plt.legend(bbox_to_anchor=(1.04, 1), loc="upper left")
plt.show()
print(i)
```

```python
%matplotlib notebook

# import required modules
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import glob

# Select all CSV files on Desktop
files = glob.glob(r"\Users\Lawrence Hsiung\NEWtests1024\rr1\*.csv")


xaxis = np.linspace(0,100, 100)
fig = plt.figure()
plt.title('NEW Characterizations coldstart ')


for file in files:
    if 'close-1.0' in file: 
        s_time = time.time()
        df = pd.read_csv(file, header = None)
        e_time = time.time()
        #print("Read time: ", (e_time-s_time), "seconds")
        df= df.T
        #print(df)
        df.columns =['Ch1', 'Ch2', 'Ch3', 'Ch4']
        zpt = sum(df['Ch3'][200:400])/len(df['Ch3'][200:400])
        df['Ch3'] = (df['Ch3']-zpt)/0.0666666 #calibration factor
        df['Ch2'] = df['Ch2'] - 0.126
        
        if len(df['Ch3']) < 10005:
            endpos = sum(df['Ch3'][8500:9500])/1000
            name = int(file[-6:-4])
            if name < 0:
                name = -name
            plt.scatter(name, endpos)
            
plt.xlabel('Run #')
plt.ylabel('End pos')

plt.show()
print("Done")
```
