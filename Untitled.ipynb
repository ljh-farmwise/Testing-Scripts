```python
filelist = []
for file in os.listdir():
    filelist.append(file)
print(filelist)
```

```python

df0 = pd.read_csv(filelist[0])
df1 = pd.read_csv(filelist[1])
df2 = pd.read_csv(filelist[2])
df3 = pd.read_csv(filelist[3])
df4 = pd.read_csv(filelist[4])
df5 = pd.read_csv(filelist[5])
df6 = pd.read_csv(filelist[6])
```

```python
df = df3
df.info()
```

```python
def starttrig(df, trig):
    start = df[df["TracDragR_N"] > trig].index[0]
    return(df[start:])
```

```python

trig = 200

ttime = df.iloc[np.argmax(df["TracDragR_N"] > trig)]["Time"] #triggers on Right channel
ittime = df[df['Time']==ttime].index.item()

plt.plot(df["Time"][ittime::], df["TracDragL_N"][ittime::]*0.224809,  label = "Left")
plt.plot(df["Time"][ittime::], df["TracDragR_N"][ittime::]*0.224809,  label = "Right")
plt.plot(df["Time"][ittime::], (df["TracDragL_N"][ittime::]+df["TracDragR_N"][ittime::])*0.224809,  label = "Added")
plt.title(filelist[6][:-4] + " Load Cell Data")
plt.xlabel("time")
plt.ylabel("lbf")
plt.legend()
```

```python
run1adds = (df0["TracDragL_N"]+df0["TracDragR_N"])*0.224809
run2adds = (df1["TracDragL_N"]+df1["TracDragR_N"])*0.224809
run3adds = (df2["TracDragL_N"]+df2["TracDragR_N"])*0.224809
run4adds = (df3["TracDragL_N"]+df3["TracDragR_N"])*0.224809
run5adds = (df4["TracDragL_N"]+df4["TracDragR_N"])*0.224809
run6adds = (df5["TracDragL_N"]+df5["TracDragR_N"])*0.224809
run7adds = (df6["TracDragL_N"]+df6["TracDragR_N"])*0.224809

run1addsRA = run1adds.rolling(window = 1000).mean()
run2addsRA = run2adds.rolling(window = 1000).mean()
run3addsRA = run3adds.rolling(window = 1000).mean()
run4addsRA = run4adds.rolling(window = 1000).mean()
run5addsRA = run5adds.rolling(window = 1000).mean()
run6addsRA = run6adds.rolling(window = 1000).mean()
run7addsRA = run7adds.rolling(window = 1000).mean()

plt.plot(range(0, len(run1addsRA[run1addsRA > 850])), run1addsRA[run1addsRA > 850], label = "Run1, 2 Act")
plt.plot(range(0, len(run2addsRA[run2addsRA > 350])), run2addsRA[run2addsRA > 350], label = "Run2, 6 Act")
plt.plot(range(0, len(run3addsRA[run3addsRA > 350])), run3addsRA[run3addsRA > 350], label = "Run3, 18 Act")
plt.plot(range(0, len(run4addsRA[run4addsRA > 350])), run4addsRA[run4addsRA > 350], label = "Run4, 18 Act")
plt.plot(range(0, len(run5addsRA[run5addsRA > 350])), run5addsRA[run5addsRA > 350], label = "Run5, 18 Act")
plt.plot(range(0, len(run6addsRA[run6addsRA > 350])), run6addsRA[run6addsRA > 350], label = "Run6, 18 Act")
plt.plot(range(0, len(run7addsRA[run7addsRA > 350])), run7addsRA[run7addsRA > 350], label = "Run7, 12 Act")


plt.title("Rolling averaging Runs, W=1000")
plt.legend()
plt.xlabel("samples")
plt.ylabel("lbf")
```

```python


for i in range(0, len(filelist)-1):
    print(filelist[i])
    df = pd.read_csv(filelist[i])
    runadds = (df["TracDragL_N"]+df["TracDragR_N"])*0.224809
    runaddsRA = runadds.rolling(window = 1000).mean()
    plt.plot(range(0, len(runaddsRA[runaddsRA > 350])), runaddsRA[runaddsRA > 350], label = filelist[i][0:8])
    plt.legend()
plt.show()
```

```python
plt.close()
lines = []
for i in range(0, len(filelist)-1):
    print(filelist[i])
    df = pd.read_csv(filelist[i])
    runadds = (df["TracDragL_N"]+df["TracDragR_N"])*0.224809
    runaddsRA = runadds.rolling(window = 1000).mean()
    
    c, b, a = np.polyfit(range(0, len(runaddsRA[runaddsRA > 850])), runaddsRA[runaddsRA > 850], deg=2)
    #lines.append(a + b * range(0, len(runaddsRA[runaddsRA > 850])))
    lines.append(a + b * range(0, len(runaddsRA[runaddsRA > 850] )) + c * np.power(range(0, (len(runaddsRA[runaddsRA > 850]))), 2))
    plt.plot(range(0, len(runaddsRA[runaddsRA > 850])), a + b * range(0, len(runaddsRA[runaddsRA > 850] )) + c * np.power(range(0, (len(runaddsRA[runaddsRA > 850]))), 2) , label = filelist[i][0:8], lw=2.5)
    
    plt.legend()
plt.title("linear fit of rolling avg data")
plt.xlabel("samples")
plt.ylabel("lbf")
plt.show()
```
